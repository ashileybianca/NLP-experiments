#Atributos léxicos

Com Spacy podemos obter diversas caracteristicas linguisticas dos tokens. Como:

Tagueamento POS (Part of Speech) / parte da frase, discurso, fala
- Classe gramatical
  - Eduardo: Substantivo
  - Foi: Verbo

- Lematizacao / forma de simplicar a frase, por ex: gato, gatos, gata, gatas é o mesmo, o que muda é plural, singular/ genero
  - desflexionar palavras
  - Gatos: Gato
  - Gatas: Gato
  - Gata: Gato

- Dependencias / como uma coisa se relaciona com a outra, relacoes e correlacoes
  - "Maria tem"
    - Maria <- Sujeito nomial <- tem
    - Alguem tem, no caso Maria

- Stopwords / palavras que nao significam nada e queremos remover
  - "As pessoas foram"
  - "As" é irrelevante

```python
 doc = nlp('Eduardo faz videos.')

print('{:10} | {:10} | {:10} | {:10} | {}'.format('Token','POS', 'LEMMA', 'DEP', 'IS_STOP'))
print('-' * 100)

for token in doc:
    print('{:10} | {:10} | {:10} |  {:10} | {}'.format(
        token.text,
        token.pos_,
        token.lemma_,
        token.dep_,
        token.is_stop
    ))
```

```python
from spacy import load

nlp = load('pt_core_news_sm')

doc = nlp('Eduardo compra passagens.')
doc = nlp('gatos cairam do muro.')

for t in doc:
  print('{:10} | {:10} | {:10} | {}'.format(
        t.text,
        t.lemma_,
        t.pos_,
        t.morph
      )
    )

# ps: mood é tipo: indicativo, conjuntivo, etc.
# verbform: finito (ação concluida), participativo, gerundio, etc.
```

Exemplo concreto

```python
from spacy import load
from spacy.matcher import Matcher

nlp = load('pt_core_news_sm')

doc = nlp(
    '''
    Maria comeu batata
    Fausto fez magia
    Allan comerá as pizzas
    Saiu um vídeo novo da Live de Python
    '''
)

matcher = Matcher(nlp.vocab)

# padrao = [
#   {'POS': 'PROPN'},
#   {'POS':'VERB'} ,
#   {'OP':'?'}, #pega só se tiver.
#   {'POS':'NOUN'}
# ]

padrao = [
    {'POS': 'PROPN'},
    #{'MORPH': {'IS_SUPERSET': ['Tense=Fut']}}
    {'MORPH': {'IS_SUPERSET': ['Tense=Past']}}
]

#'MORPH': {'IS_SUPERSET': ['Gender=Fem']}

matcher.add('Regra A', [padrao])

matches = matcher(doc)

for id_, start, stop in matches:
  print(doc[start:stop].text)
```
